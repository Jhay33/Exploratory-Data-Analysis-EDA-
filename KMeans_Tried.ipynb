```python
# K-Means for xclara dataset
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('xclara.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
f1 = data['V1'].values
f2 = data['V2'].values
X = np.array(list(zip(f1,f2)))
print(X)
```

```python
from sklearn.cluster import KMeans

# no.of clusters

kmeans = KMeans(n_clusters = 3)

# fitting the i/p data

kmeans = kmeans.fit(X)

# getting the cluster labels

labels = kmeans.predict(X)
```

```python
 # centroid values

centroids = kmeans.cluster_centers_

print('centroid values :')
print(centroids)
print()
print('labels :',labels)
```

KMeans for Beer Dataset

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_beer = pd.read_csv('beer.txt', sep=' ')
```

```python
data_beer
```

```python
X = data_beer.drop('name', axis=1)
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3, random_state = 100)

model.fit(X)
```

```python
model.labels_
```

```python
data_beer['cluster'] = model.labels_
```

```python
data_beer = data_beer.sort_values(by = 'cluster')
```

```python
data_beer
```

```python
model.cluster_centers_
```

```python
plt.scatter(X.calories[:], X.alcohol[:], s=25)
plt.xlabel('Calories')
plt.ylabel('Alcohol')
```

```python
# scale the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
model_1 = KMeans(n_clusters=3,random_state=100)

model_1.fit(X_scaled)
```

```python
data_beer['NewCluster'] = model_1.labels_
```

```python
data_beer.sort_values(by = 'NewCluster')
```

```python
from sklearn import metrics
metrics.silhouette_score(X_scaled, model_1.labels_)
```

```python
scores = []

for i in range(2,10):
  model = KMeans(n_clusters = i, random_state=100)
  model.fit(X_scaled)
  scores.append(metrics.silhouette_score(X_scaled,model.labels_))
```

```python
scores
```

```python
k_range = list(range(2,10))
plt.plot(k_range,scores)
plt.grid(True)
plt.xlabel('cluster')
plt.ylabel('silhouette')
```
