```python
energy_median = Data['Energy Consumed (kWh)'].median()
charging_rate_median = Data['Charging Rate (kW)'].median()
distance_driven_median = Data['Distance Driven (since last charge) (km)'].median()

Data['Energy Consumed (kWh)'].fillna(energy_median, inplace=True)
Data['Charging Rate (kW)'].fillna(charging_rate_median, inplace=True)
Data['Distance Driven (since last charge) (km)'].fillna(distance_driven_median, inplace=True)
```

```python
from sklearn.linear_model import LinearRegression

X = Data[['Energy Consumed (kWh)', 'Charging Duration (hours)']].dropna()
y = Data['Charging Cost (USD)']

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

print("Effect of Energy Consumed on Charging Cost:", model.coef_[0])
print("Effect of Charging Duration on Charging Cost:", model.coef_[1])
```
